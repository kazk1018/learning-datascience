### カラムとSeries

### 特定のデータを取得する
* table[array(boolean)]でtrueのデータだけ取得できる
* tabel.species=='setosa'はspeciesがsetosaならtrue, そうでなければfalseを返す

```python
table[table.species=='setosa']
```

```python
table[table.sepal_length==4.8]
```

## 基礎統計量
### 代表値
* 平均・・・全データの平均
* モード・・・最頻値、一番出現した値
* メディアン・・・中央値、全データ数の半分の点にある値

```python
table.sepal_length.mean()
```

```python
table.sepal_length.mode()
```

```python
table.sepal_length.median()
```

### 散らばり
* 分散・・・値ー平均の二乗平均和
* 標準偏差・・・分散の平方根

```python
table.sepal_length.var()
```

```python
table.sepal_length.std()
```

```python
table.sepal_length.std()**2 # 二乗
```
